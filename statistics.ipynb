```python
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
```

```python
data = np.random.normal(loc=100,scale=20,size=500).astype(int)
collection = Counter(data)
```

```python
xs=range(data.min(),data.max()+1)
ys=[collection[x] for x in xs]
plt.bar(xs,ys)
plt.show()
```

```python
# Find median

def median(data):
    n=len(data)
    data=sorted(data)
    midpoint=n//2
    if (n%2==0):
        li=midpoint-1
        hi=midpoint
        return (data[li]+data[hi])/2
    else:
        return data[midpoint]

```

```python
# Find quantile (Value x in the sorted series where p part of the values are .< x)

def quantile(x,p):
    x=sorted(x)
    i=int(len(x)*p)
    return x[i]

quantile(data,0.4)
```

```python
# Find mean
def mean(x):
    return sum(data)/len(data)

# Find variance (Variance is a measure of dispersion of the data)
def variance(x):
    x_bar = mean(x)
    return round(sum([(i-x_bar)**2 for i in x]) / len(x), 6)

def de_mean(x):
    """translate x by subtracting its mean (so the result      has mean 0)"""    
    x_bar = mean(x)
    return [x_i - x_bar for x_i in x]
```

```python
# Find Covariance (measure of correlation between variables)
from vectors import dot
import math
def covariance(x,y):
    return dot(de_mean(x), de_mean(y))/(len(x)-1)

# Find unitless measure of correlation
def correlation(x,y):
    stdev_x = math.sqrt(variance(x))
    stdev_y=math.sqrt(variance(y))
    if (stdev_y>0 and stdev_x>0):
        return (covariance(x,y) / stdev_x ) / stdev_y
    else: return 0
```
